```python
import numpy as np
import pandas as pd

df = pd.read_csv("merged_ss.csv")
```

```python
# Preview Data

df.head()
```

```python
# Drop columns with least signifigance to hypothesis
```

```python
# Drop Rows with null values

# df.dropna()

df.isnull().sum(axis = 0)
```

```python
# Add Profit Margin Column

df['Profit Margin Percentage']=df.apply(lambda x: x.Profit/x.Sales if x.Profit!=0 else 0, axis=1) * 100

df['Shipping_Cost'] = pd.to_numeric(df['Shipping_Cost'])

df.head()
```

```python
# Convert Order Date and Ship Date to date/time

df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
```

```python
# create Days to Ship (ship date - order date)

df['Days to Ship'] = (df['Ship_Date'] - df['Order_Date']).dt.days

df['Days to Ship']
```

```python
df.tail(100)
```

```python
# Raw Average Sale

df["Sales"].mean()
```

```python
# Average Sale Price by Category

df[["Category", "Sales"]].groupby("Category").mean()
```

```python
# Average Sales Price by Sub_Category

df.groupby(['Category', 'Sub_Category'])['Sales'].mean()
```

```python
#df.groupby(['Category', 'Sub_Category'])['Profit Margin Percentage'].mean()

df.groupby(['Category','Sub_Category'])['Profit Margin Percentage'].mean()
```

```python
# Sales totals by Sub_Category

df.groupby(['Category', 'Sub_Category'])['Sales'].sum()
```

```python
# Total Sales by Country

df.groupby(['Country'])['Sales'].sum()
```

```python
# Count for each Order Priority

df[['Order Priority', 'Ship_Mode']].groupby('Order Priority').count()
```

```python
# Average Profit Margin by Order Priority

df[['Order Priority', 'Profit Margin Percentage']].groupby('Order Priority').mean()
```

```python
# Order Priority (Count AND Profit Margin) within each Ship Mode

# df.pivot_table(index='Order Priority',columns='Ship Mode',aggfunc='size',fill_value=0).unstack()

df.groupby(['Ship_Mode','Order Priority'])['Profit Margin Percentage'].agg(['count','mean'])#.reset_index()
       
```
