```python
# drop columns as necessary
super_df_drop = super_df.drop(super_df.columns.difference(['State','Country','Category','Sub-Category', 'Sales', 'Profit','Order Priority', 'Discount', 'Quantity']), 1)
super_df_drop.head()
```

```python
# look at data types
super_df_drop.dtypes
```

```python
# fix sales data type
super_df_drop['Sales'] = super_df_drop['Sales'].str.replace(',', '')
super_df_drop['Sales'] = super_df_drop['Sales'].str.replace('$', '')
super_df_drop['Sales'] = super_df_drop['Sales'].astype(float)
```

```python
# fix profit data type
super_df_drop['Profit'] = super_df_drop['Profit'].str.replace(',', '')
super_df_drop['Profit'] = super_df_drop['Profit'].str.replace('$', '')
super_df_drop['Profit'] = super_df_drop['Profit'].astype(float)
```
