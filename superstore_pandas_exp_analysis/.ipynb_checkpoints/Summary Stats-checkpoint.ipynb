```python
import numpy as np
import pandas as pd

df = pd.read_csv("merged_ss.csv")
```

```python
# Preview Data

df.head()
```

```python
# Drop columns with least signifigance to hypothesis
```

```python
# Drop Rows with null values

# df.dropna()

df.isnull().sum(axis = 0)
```

```python
# Add Profit Margin Column

df['Profit Margin Percentage']=df.apply(lambda x: x.Profit/x.Sales if x.Profit!=0 else 0, axis=1) * 100

df['Shipping_Cost'] = pd.to_numeric(df['Shipping_Cost'])

df.head()
```

```python
# Convert Order Date and Ship Date to date/time

df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
```

```python
# create Days to Ship (ship date - order date)

df['Days to Ship'] = (df['Ship_Date'] - df['Order_Date']).dt.days

df['Days to Ship']
```

```python
df.tail(100)
```

```python
# Raw Average Sale

df["Sales"].mean()
```

```python
# Average Sale Price by Category

df[["Category", "Sales"]].groupby("Category").mean()
```

```python
# Average Sales Price by Sub_Category

df.groupby(['Category', 'Sub_Category'])['Sales'].mean()
```

```python
#df.groupby(['Category', 'Sub_Category'])['Profit Margin Percentage'].mean()

df.groupby(['Category','Sub_Category'])['Profit Margin Percentage'].mean()
```

```python
# Sales totals by Sub_Category
df_sales_grouped=df.groupby(['Category', 'Sub_Category'])['Sales'].sum().reset_index()
df_sales_grouped
```

```python
# Total Sales by Country

df.groupby(['Country'])['Sales'].sum()
df.sort_values(by = 'Sales', ascending = False )
```

```python
# Count for each Order Priority

df[['Order Priority', 'Ship_Mode']].groupby('Order Priority').count()
```

```python
# Average Profit Margin by Order Priority

df[['Order Priority', 'Profit Margin Percentage']].groupby('Order Priority').mean()
```

```python
# Order Priority (Count, Profit Margin, Dollars) within each Ship Mode

df.groupby(['Ship_Mode','Order Priority'])['Profit Margin Percentage'].agg(['count','mean','sum'])#.reset_index()
       
```

```python
# change returned "yes" to 1

df = df.replace('Yes', 1)
df.head()
```

```python
# fill Returned NaNs with 0

df['Returned'] = df['Returned'].fillna(0)
df.head(4)
```

```python
df.tail(4)
```

```python
# Get Count of Returns

df.value_counts('Returned')
```

```python
# Returns total Dollars by Category and Subcategory

# df.groupby(['Category', 'Sub_Category'])['Returned'].sum()
# total dollar of return :

#amt_return=df.loc[df.Returned==1].groupby(['Category', 'Sub_Category','Country'])[['Sales']].sum().reset_index()

amt_return=df.loc[df.Returned==1].groupby(['Country'])[['Category','Sub_Category','Sales']].sum().reset_index()
amt_return.rename(columns={ 'Sales':'Returned_amt'}, inplace=True)

amt_return.sort_values(by = 'Returned_amt', ascending = False ).head()

```

```python
# Returns total Dollars by Category and Subcategory


# df.groupby(['Category', 'Sub_Category'])['Returned'].sum()
# total dollar of return :
dollar_amt_return=df.loc[df.Returned==1].groupby(['Category', 'Sub_Category'])[['Country','Sales']].sum().reset_index()
dollar_amt_return.rename(columns={ 'Sales':'Returned_amt'}, inplace=True)

dollar_amt_return.head()
```

```python
dollar_amt_return.columns
```

```python
# join two dataframes dollar_amt_return and df_sales_grouped
df_combined = pd.merge( df_sales_grouped,dollar_amt_return, left_on = ['Category', 'Sub_Category', 'Country'], right_on=['Category', 'Sub_Category', 'Country'] 
         , how ='left')

df_combined['return_rate'] = round((df_combined.Returned_amt/df_combined.Sales)*100,2)

print(df_combined.shape)

df_combined.sort_values(by = 'return_rate', ascending = False )

```

```python
# Average return Profit Margin Percent
df.loc[df.Returned==1].groupby(['Category', 'Sub_Category'])['Profit Margin Percentage'].mean()
```

```python
# Dollars in Profit Lost to Returns by Subcategory
df.loc[df.Returned==1].groupby(['Category', 'Sub_Category'])['Profit Margin Percentage'].sum()
```
