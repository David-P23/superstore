```python
# Clean the data and remove unneccesary columns
# To do: 

# Drop Cloumns
# Date Modification
# Price Correction


import datetime as dt
import numpy as np
import pandas as pd

#  Import and read the file:
super_df = pd.read_csv("resources/Global Superstore Orders 2016.csv")
super_df.head()
```

```python
# fix sales data type
super_df['Sales'] = super_df['Sales'].str.replace(',', '')
super_df['Sales'] = super_df['Sales'].str.replace('$', '')
super_df['Sales'] = super_df['Sales'].astype(float)
```

```python
# fix profit data type
super_df['Profit'] = super_df['Profit'].str.replace(',', '')
super_df['Profit'] = super_df['Profit'].str.replace('$', '')
super_df['Profit'] = super_df['Profit'].astype(float)
```

```python
# fix date time dtypes in ship date and order date

super_df['Order Date']= pd.to_datetime(super_df['Order Date'])
super_df['Ship Date']= pd.to_datetime(super_df['Ship Date'])
```

```python
super_df.head()
```

```python
#check data types
super_df.dtypes
```

```python
# rename columns with spaces for easier manipulation

# df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'}, inplace=True)
super_df.rename(columns={'Sub-Category': 'Sub_Category', 'Shipping Cost':'Shipping_Cost', 'Order Date':'Order_Date','Ship Date': 'Ship_Date', 'Ship Mode':'Ship_Mode', 'Product Name':'Product_Name'}, inplace=True)
super_df.columns
```

```python
# drop columns 
# Row ID,Order ID,Customer ID,Customer Name,Postal Code,Product ID all dropped
df_drop = super_df.drop(['Row ID','Order ID','Customer ID','Customer Name','Postal Code','Product ID'], 1)
df_drop.head()
```

```python
# put cleaned dataframe into new csv in resources folder for further analysis
df_drop.to_csv("resources/cleaned_superstore_data.csv")
```
