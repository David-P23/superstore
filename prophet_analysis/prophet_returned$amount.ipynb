```python
# import dependencies
import pandas as pd
from prophet import Prophet
```

## All Orders

```python
# import Superstore (all orders)
super_df = pd.read_csv('../resources/superstore.csv')
super_df.head()
```

```python
# Create a new df with only the order date & returned $ amount - then rename to ds and y, respectively
returned_amt = super_df[['Order Date', 'Returned $ Amount']]
returned_amt = returned_amt.rename(columns={'Order Date': 'ds', 'Returned $ Amount': 'y'})
returned_amt.head()
```

```python
# initiate new Prophet object
m = Prophet()
```

```python
#call Prophet fit method and pass in the returned_amt df
m.fit(returned_amt)
```

```python
# extend our date range into the future by 365 days
# includes all current date and extends our data through end of 2016
future_returned_amt = m.make_future_dataframe(periods=365)
future_returned_amt.tail()
```

```python
# create predicted returned$ values (yhat) as well as lower & upper values for uncertainty intervals
forecast_returned_amt = m.predict(future_returned_amt)
forecast_returned_amt[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# plot the predicted returned$ values
pred_returned_amt = m.plot(forecast_returned_amt)
```

```python
# plot the forecast components - overall trend, yearly seasonality, & weekly seasonality
pred_returned_amt_components = m.plot_components(forecast_returned_amt)
```
