```python
# import dependencies
import pandas as pd
from prophet import Prophet
```

## All Orders

```python
# import Superstore (all orders)
super_df = pd.read_csv('../resources/superstore.csv')
super_df.head()
```

```python
# Create a new df with only the order date & profit $ amount - then rename to ds and y, respectively
profit = super_df[['Order Date', 'Profit']]
profit = profit.rename(columns={'Order Date': 'ds', 'Profit': 'y'})
profit.head()
```

```python
# initiate new Prophet object
m = Prophet()
```

```python
#call Prophet fit method and pass in the profit df
m.fit(profit)
```

```python
# extend our date range into the future by 365 days
# includes all current date and extends our data through end of 2016
future_profit = m.make_future_dataframe(periods=365)
future_profit.tail()
```

```python
# create predicted profit values (yhat) as well as lower & upper values for uncertainty intervals
forecast_profit = m.predict(future_profit)
forecast_profit[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# plot the predicted profit values
pred_profit = m.plot(forecast_profit)
```

```python
# plot the forecast components - overall trend, yearly seasonality, & weekly seasonality
pred_profit_components = m.plot_components(forecast_profit)
```
