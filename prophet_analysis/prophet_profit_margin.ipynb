```python
# import dependencies
import pandas as pd
from prophet import Prophet
```

## All Orders

```python
# import Superstore (all orders)
super_df = pd.read_csv('../resources/superstore.csv')
super_df.head()
```

```python
# Create a new df with only the order date & profit margin - then rename to ds and y, respectively
profit_marg = super_df[['Order Date', 'Profit Margin Percentage']]
profit_marg = profit_marg.rename(columns={'Order Date': 'ds', 'Profit Margin Percentage': 'y'})
profit_marg.head()
```

```python
# initiate new Prophet object
m = Prophet()
```

```python
#call Prophet fit method and pass in the profit_marg df
m.fit(profit_marg)
```

```python
# extend our date range into the future by 365 days
# includes all current date and extends our data through end of 2016
future_profit_marg = m.make_future_dataframe(periods=365)
future_profit_marg.tail()
```

```python
# create predicted profit margin values (yhat) as well as lower & upper values for uncertainty intervals
forecast_profit_marg = m.predict(future_profit_marg)
forecast_profit_marg[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# plot the predicted profit margin values
pred_profit_marg = m.plot(forecast_profit_marg)
```

```python
# plot the forecast components - overall trend, yearly seasonality, & weekly seasonality
pred_profit_marg_components = m.plot_components(forecast_profit_marg)
```
